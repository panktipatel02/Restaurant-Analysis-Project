# **Task 1: Top Cuisines**

Determine the top three most
common cuisines in the dataset.

Calculate the percentage of
restaurants that serve each of the top
cuisines.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read File
level1_dataset_df = pd.read_csv('/content/Dataset .csv')
level1_dataset_df.head()
```

```python
level1_dataset_df.describe()
```

```python
level1_dataset_df.info()
```

```python
level1_dataset_df.shape
```

```python
level1_dataset_df.Cuisines.unique()
```

```python
top_three_cuisines = level1_dataset_df.Cuisines.value_counts()
top_three_cuisines
```

```python
# top three most common cuisines
plt.figure(figsize=(10, 6))
sns.barplot(x=top_three_cuisines.head(3).index, y=top_three_cuisines.head(3).values, palette='crest')
plt.title('Top 3 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
```

```python
split = level1_dataset_df['Cuisines'].str.split(',').explode()
cuisine_counts =split.value_counts()
total_restaurants = len(level1_dataset_df)
cuisine_percentages = (cuisine_counts / total_restaurants) * 100
print("Top Cuisines and Their Percentage of Restaurants:")
cuisine_percentages.head(10)
```

# **Task 2: City Analysis**

Identify the city with the highest number
of restaurants in the dataset.

Calculate the average rating for
restaurants in each city.

Determine the city with the highest
average rating.

```python
city_df = level1_dataset_df.City.unique()
```

```python
#Identify the city with the highest number of restaurants in the dataset.
highest_restaurants = level1_dataset_df['City'].value_counts().idxmax()
print(f"City with the highest number of restaurants: {highest_restaurants}")
```

```python
# Calculate the average rating for restaurants in each city
average_rating = level1_dataset_df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
average_rating
```

```python
#city with the highest average rating
highest_average_rating_city = average_rating.idxmax()
highest_average_rating_city
```

# **Task 3: Price Range Distribution**

Create a histogram or bar chart to
visualize the distribution of price ranges
among the restaurants.

Calculate the percentage of restaurants
in each price range category.

```python
price_range_df = level1_dataset_df['Price range']
```

```python
price_range_counts = level1_dataset_df['Price range'].value_counts().sort_index()
price_range_counts
```

```python
#Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(price_range_df, bins=4, edgecolor='black', range=(1, 4))
plt.xlabel('Price Range Category')
plt.ylabel('Number of Restaurants')
plt.title('Histogram of Price Ranges Among Restaurants')
plt.xticks([1, 2, 3, 4], ['1 (Inexpensive)', '2 (Moderate)', '3 (Expensive)', '4 (Very Expensive)'])
plt.show()
```

```python
#Calculate the percentage of restaurants in each price range category.
total_restaurants = len(level1_dataset_df)
price_range_percentages = (price_range_counts / total_restaurants) * 100
price_range_percentages
```

# **Task 4: Online Delivery**

Determine the percentage of restaurants
that offer online delivery.

Compare the average ratings of restaurants
with and without online delivery.

```python
#Determine the percentage of restaurants that offer online delivery.
online_delivery_percentage = (level1_dataset_df['Has Online delivery'] == 'Yes').mean() * 100
online_delivery_percentage
```

```python
#Compare the average ratings of restaurants with and without online delivery.
delivery_rating = level1_dataset_df.groupby('Has Online delivery')['Aggregate rating'].mean()
delivery_rating
```
